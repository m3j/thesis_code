#### Defining which genomes are only present in one region. 

```python
genome_country = meta_binary_sample[all_genomes]
```

```python
meta_binary_sample.head(5)
```

```python
genome_country['region'] =  meta_binary_sample['region']
```

```python
genome_country.head()
```

```python
sample_in_country = {}

for genome in all_genomes:
    all_regions = []
    for index, row in genome_country.iterrows():
        if row[genome] == 1 :
            all_regions.append(row['region'])
 
    sample_in_country[genome] = set(all_regions)
print(sample_in_country)
```

```python
present_only_europe = []
present_only_china = []
present_both = []

for key, value in sample_in_country.items():
    if value == {'europe'}:
        present_only_europe.append(key)
    elif value == {'china'}:
        present_only_china.append(key)
    else:
        present_both.append(key)
        
```

```python
print(len(present_both))
print(len(present_only_china))
print(len(present_only_europe))
```

```python
present_both_short = [ '.'.join(x.split('.')[:1]) for x in present_both]
present_only_china_short = [ '.'.join(x.split('.')[:1]) for x in present_only_china]
present_only_europe_short = [ '.'.join(x.split('.')[:1]) for x in present_only_europe]
```

### listing the genomes based on  the number  of appearences of each genome.  

```python
smallest_presence = []
small_presence = []
big_presence = []
biggest_presence = []

for j in range(0,57,1):
    if genome_country.drop('region', axis=1).sum()[j] <= 5:
        smallest_presence.append(genome_country.drop('region', axis=1).sum().index[j])
    elif genome_country.drop('region', axis=1).sum()[j] > 5 and genome_country.drop('region', axis=1).sum()[j] <= 15:
        small_presence.append(genome_country.drop('region', axis=1).sum().index[j])
    elif genome_country.drop('region', axis=1).sum()[j] > 15 and genome_country.drop('region', axis=1).sum()[j] <= 30:
        big_presence.append(genome_country.drop('region', axis=1).sum().index[j])
    else:
        biggest_presence.append(genome_country.drop('region', axis=1).sum().index[j])
```

```python
print(len(small_presence))
print(len(smallest_presence))
print(len(big_presence))
print(len(biggest_presence))
```
